```python
import pandas as pd
```

```python
df2 = pd.read_csv('ECG200_TRAIN.csv', delimiter="  ", header=None)
df2
```

```python
df2[0].value_counts()
```

```python
df2.sort_values(by=[0], inplace=True)
```

```python
pred1 = df2.iloc[:31]
pred2 = df2.iloc[31:]

fold = []

for i,j in zip(range(0,30,3), range(0,70,7)):
    fold.append(pd.concat([pred1.iloc[i:i+3],pred2.iloc[j:j+7]]) )

    
fold[8]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics

Avg_accuracy = 0


test = []
train = []
for i in range(0,10):

    list1 =[]
    test.append(fold[i])
    
    for j in range(0,10):
        
        if j != i:
            list1.append(fold[j])
    
    train.append(pd.concat(list1))
    
    clf=RandomForestClassifier(n_estimators=100)
    
    Y_train = train[i].iloc[:,0]
    X_train = train[i].iloc[:,1:]

    clf.fit(X_train,Y_train)
    
    #testing data set
    
    Y_test = test[i].iloc[:,0]
    X_test = test[i].iloc[:,1:]

    
    y_pred=clf.predict(X_test)
        
    print("Accuracy:",metrics.accuracy_score(Y_test, y_pred))
    
    Avg_accuracy += metrics.accuracy_score(Y_test, y_pred)

Avg_accuracy = Avg_accuracy/10

print("\nAverage accuracy is: ", Avg_accuracy)                                 
```
